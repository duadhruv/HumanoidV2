```python
import nltk
from collections import  defaultdict
#nltk.download('punkt')
from nltk.tokenize import word_tokenize, sent_tokenize

text = open('3.txt', 'r').read()
sentences = sent_tokenize(text)
print("total sentences are  : " ,len(sentences))


test = sentences[:1]
train = sentences[1:]
```

```python


def getFilteredWords(sentences):
    """ Filter words and print words and vocabalary length """
    words = []
    for sentence in sentences:
      sent_words = word_tokenize(sentence)
      words += sent_words

    bad_words = ['.','#',',',"'"]
    for word in words:
      if word in bad_words:
        words.remove(word)

    vocabalary = set(words)
    return words,vocabalary

def printCountProbab(freqWords, total_words , model_name):
    print()
    print("Model : " , model_name)
    print("frequencies of words is : ")
    print(freqWords)

    probability = {word:freq/total_words for word,freq in freqWords.items()}
    print()
    print("probabilty of words is : ")
    print(probability)
    return probability

def printCountAProbab(words,vocabalary,corpus_size):
    freqWords = defaultdict(int)
    for word in words:
      freqWords[word] += 1
    for word in vocabalary:
      temp = freqWords[word]

    freqProb = printCountProbab(freqWords,corpus_size,"unsmoothed")

    add1FreqWords = defaultdict(int)
    for word in vocabalary:
      add1FreqWords[word] = freqWords[word] + 1

    add1Prob = printCountProbab(add1FreqWords,corpus_size + len(vocabalary),"Add 1 Smoothing")
    
    Nc = defaultdict(int)
    for key,val in freqWords.items():
      Nc[val] += 1

    turingFreqWords = defaultdict(int)  
    for word in vocabalary:
      turingFreqWords[word] = (freqWords[word] + 1) * Nc[freqWords[word] + 1] / Nc[freqWords[word]] 
    turingProb = printCountProbab(turingFreqWords, corpus_size , "Good Turing smoothing")

    return freqProb,add1Prob,turingProb


def printProbab(words,prob,model_name):
    print()
    print("Test Words : ")
    sent_prob = 1
    for word in words:
      print(word + ": " , prob[word], end = " ,")
      sent_prob *= prob[word]  
    print()
    print("Total Senetecnce Probability for " + model_name + " is : ",sent_prob)

def findSentenceProbability(sentence,freqProb,add1Prob,turingProb):
    words,vocab = getFilteredWords(sentence)
    printProbab(words,freqProb,"Without smoothing" )
    printProbab(words,add1Prob,"With Add-1 smoothing")
    printProbab(words,turingProb,"With Good Turing Smoothing")

vocab = set()
trainWords,trainVocab = getFilteredWords(train)
testWords,testVocab = getFilteredWords(test)
vocab = trainVocab.union(testVocab)
print("Total words is : ",len(trainWords) + len(testWords))
print("Total unique words is : " ,len(vocab))
freqProb, add1Prob, turingProb = printCountAProbab(trainWords,vocab,len(trainWords) + len(testWords))
words,vocab = getFilteredWords(test)
printProbab(words,freqProb,"unsmoothed" )
printProbab(words,add1Prob,"Add 1 Smoothing")
printProbab(words,turingProb,"Good Turing smoothing")
```
