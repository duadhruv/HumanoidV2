`````````python
import re
import collections
import pandas


with open('/content/text.txt', 'r') as f:
    input = f.read()
print("Input: ",input,"\n")
words = input.split()
word_freq = collections.defaultdict(int)

for word in words:
    word_freq[' '.join(word) + ' _'] += 1
    
print("Vocabulary size: ",len(word_freq),"\n")
print('freq\tword')
for i in word_freq:
    print(str(word_freq[i])+'\t'+str(i))
`````````

`````````python
char_freq = collections.defaultdict(int)
for word, freq in word_freq.items():
    chars = word.split()
    for char in chars:
        char_freq[char] += freq
print('char\tfreq')
for i in char_freq:
    print(str(char_freq[i])+'\t'+str(i))
`````````

`````````python
def merge_pairs(best_pair, word_freq):
    merged_dict = {}
    bigram = re.escape(' '.join(best_pair))
    p = re.compile(r'(?<!\S)' + bigram + r'(?!\S)')
    for word in word_freq:
        w_out = p.sub(''.join(best_pair), word)
        merged_dict[w_out] = word_freq[word]
    return merged_dict

vocab = list()
for i in char_freq:
    vocab.append(i)

for i in range(40):
    pairs = collections.defaultdict(int)
    for word, freq in word_freq.items():
        chars = word.split()
        for i in range(len(chars)-1):
            pairs[chars[i], chars[i+1]] = freq
    best = max(pairs, key=pairs.get)
    vocab.append(''.join(best))
    print("Iteration :", i)
    word_freq = merge_pairs(best, word_freq)
    
    subword = collections.defaultdict(int)
    for word, freq in word_freq.items():
        chars = word.split()
        for char in chars:
            subword[char] += freq

    print(best)
    print("Total subwords ",len(subword),"\n")
    for i in subword:
        print(str(subword[i])+'\t'+str(i))
    print("````````")

print("Vocabulary: ",vocab)
`````````
