```python
from time import time
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
import pandas as pd

input = []
with open('dataset.csv','r') as f:
  a = f.readlines()
  input = [[_.lower() for _ in l.split(',')] for l in a]
print(input)
trainX = [i[1] for i in input[20:]]
trainY = [i[0] for i in input[20:]]
testX = [i[1] for i in input[1:20]]
testY = [i[0] for i in input[1:20]]
print(trainX)
```

```python
TFVectorizer = CountVectorizer()
TrainTFX = TFVectorizer.fit_transform(trainX)
```

```python
TestTFX = TFVectorizer.transform(testX)
NB = MultinomialNB()
NB.fit(TrainTFX, trainY)
predY = NB.predict(TestTFX)
```

```python
print(predY)
print(testY)
```

```python
score1 = metrics.accuracy_score(testY, predY)
print("Accuracy : " + str(score1))
```

```python
print(metrics.classification_report(testY, predY,target_names=['ham', 'spam']))
```
